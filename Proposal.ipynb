```python
#extract year , month and date
def get_year_mon_day(date):
    date= str(date) # convert to string 
    splited= date.split("-") # to get month, day, and year and time
    year= splited[0] # get year 
    month=splited[1]
    day= splited[2][:2] # extract only the day
    return year, month ,day
```

```python
def get_year(date):
    date= str(date) # convert to string 
    splited= date.split("-") # to get month, day, and year and time
    year= splited[0] # get year 
    return year

def get_month(date):
    date= str(date) # convert to string 
    splited= date.split("-") # to get month, day, and year and time
    month= splited[1] # get year 
    return month
def get_day(date):
    date= str(date) # convert to string 
    splited= date.split("-") # to get month, day, and year and time
    day= splited[2][:2] # extract only the day, month ,day
    return day
```
